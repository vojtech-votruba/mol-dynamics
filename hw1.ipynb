### Deriving the Contour Expression

Given the Hamiltonian $H(p,q) = \frac{p^2}{2m} + U(q)$, we can easily see that a phase space contour satisfying the constant energy $E$ constraint can be obtained by first rearranging the terms
$$
p^2 = 2m(E - U(q)),
$$
and then solving the quadratic equation above
$$
p = \pm\sqrt{2m(E-U(q))},
$$
with this we got two branches of solutions, one corresponding to positive $p$ and the other to negative $p$. 

### Harmonic Oscillator
$U(q) = \frac{1}{2}kq^2$

```python
import numpy as np
from matplotlib import pyplot as plt

m = 1.0
k = 1.0
eps = 1e-7

fig, ax = plt.subplots()
ax.set_ylabel("p")
ax.set_xlabel("q")
ax.set_aspect(1)
ax.set_title("Harmonic Oscillator Contour Plot")

energies = np.linspace(eps,2,11)

for E in energies:
    q = np.linspace(-np.sqrt(2*E/k) + eps, np.sqrt(2*E/k) - eps, 500)
    
    p_pos = np.sqrt(2*m*(E - 0.5 * k * q**2))
    p_neg = -p_pos

    ax.plot(q,p_pos, color="teal")
    ax.plot(q,p_neg, color="teal")

plt.show()
```

### Double Well Potential
$U(q) = \frac{D_0}{a^4}(q^2-a^2)^2$

```python
D0 = 1.0
a = 1.0

fig, ax = plt.subplots()
ax.set_ylabel("p")
ax.set_xlabel("q")
ax.set_title("Double Well Contour Plot")
ax.set_aspect(1)

energies = np.linspace(eps,2,11)

for E in energies:
    q = np.linspace(-np.sqrt(a**2 + np.sqrt(a**4 * E/D0)) + eps, np.sqrt(a**2 + np.sqrt(a**4 * E/D0)) - eps, 500)

    p_pos = np.sqrt(2*m*(E - D0/a**4 * (q**2 - a**2)**2))
    p_neg = -p_pos

    ax.plot(q,p_pos, color="teal")
    ax.plot(q,p_neg, color="teal")

plt.show()
```
