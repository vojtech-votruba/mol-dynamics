Using the well-known 1D canonical distribution function, we can write 
$$\rho(p,q) \sim \exp\left(-\beta H(p,q)\right) = \exp\left(-\beta\frac{p^2}{2m} - \beta U(q)\right), \quad \beta = \frac{1}{k_B T}.$$
Specifically, for the double well potential
$$\rho(p,q) \sim \exp\left(-\beta\frac{p^2}{2m} -\beta \frac{D_0}{a^4}(q^2 - a^2)^2\right).$$
Since we are in classical physics, we can split the exponential into two parts, the part dependent on $q$ and the part dependent on $p$. The assignment asks for the probability density as the function of position, so we omit the momentum part. Finally, we get
$$\rho(q;T) \sim \exp\left(-\frac{1}{k_B T} \frac{D_0}{a^4}(q^2 - a^2)^2\right).$$

```python
import numpy as np
from matplotlib import pyplot as plt

q = np.linspace(-1, 1, 500)
kB = 1.0
T = 10.0
D0 = 2.0
a = 0.5

plt.style.use("classic")
```

```python
# Potential energy
U = D0 / a**4 * (q**2 - a**2) ** 2

plt.xlabel("q")
plt.ylabel("U(q)")
plt.title("Potential energy")
plt.plot(q,U)

plt.show()
```

```python
# Probability density
rho = np.exp(- 1/(kB * T) * U)
Z = np.trapezoid(rho)

rho /= Z

plt.xlabel("q")
plt.ylabel("ρ")
plt.title(f"Probability density at T = {T}")
plt.plot(q,rho)

plt.show()
```
