### Analytic Solution

Let us start with the Hamilton's equations of motion in the general form

$$
\dot{q} = \frac{\partial H}{\partial p},\quad \dot{p} = -\frac{\partial H}{\partial q}.
$$

Plugging in the given Hamiltonian $H = \frac{p^2}{2m} + \frac{1}{2}kq^2$, the two equations become

$$
\dot{q} = \frac{p}{m}, \quad \dot{p} = -kq.
$$

This is essentially a system of first-order ODE's. Systems like this are usually solved by using the matrix exponential. Since this approach is a bit tiresome, we will try another way which consists of solving a one second-order ODE instead of solving two first-order ODE's. Taking the derivative of the first equation we get
$$
\ddot{q} = \frac{\dot{p}}{m} = -\frac{k}{m}q
$$
where we plugged in $\dot{p}$ from the expression above. The resulting equation is homogenious, so we can immediately see that the solution is going to be in the form
$$
q(t) = A\sin(\omega t) + B\cos(-\omega t), \quad \omega = \sqrt{k/m},
$$
and similarly
$$
p(t) = Am\omega\cos(\omega t) + Bm\omega\sin(-\omega t).
$$
Plugging in the initial conditions $q(0) = 0.65$, $p(0) = 1.0$, and the given parameters $m=1.15$, $k=1.25$, we finally get
$$
q(t) =  0.83\sin(1,04 t)+ 0.65\cos(-1,04t), \\
p(t) = \cos(1,04 t)+ 0.78\sin(-1,04t)
$$

### Time Dependence

```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use("bmh")

m = 1.15
k = 1.25
omega = np.sqrt(k/m)
A = 1/(m*omega)
B = 0.65

t = np.linspace(-10,10,500)

q = A*np.sin(omega*t) + B*np.cos(-omega*t)
p = A*m*omega*np.cos(omega*t) + B*m*omega*np.sin(-omega*t)
H = p**2/(2*m) + 0.5*k*q**2 

fig, ax = plt.subplots()
ax.set_xlabel("t")
ax.set_ylabel("q,p")

ax.plot(t, q, label="q")
ax.plot(t, p, label="p")
ax.plot(t, H, label="H")
ax.legend()

plt.show()
```
