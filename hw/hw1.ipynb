### Deriving the Contour Expression

Given the Hamiltonian $H(p,q) = \frac{p^2}{2m} + U(q)$, we can easily see that the equation for a contour with a set energy $E$ can be obtained by first rearranging the terms
$$
p^2 = 2m(E - U(q)),
$$
and solving the quadratic equation
$$
p = \pm\sqrt{2m(E-U(q))}.
$$
As we can see, the solution has two branches, one corresponds to positive a $p$ and the other one to a negative $p$. 

### Harmonic Oscillator
$U(q) = \frac{1}{2}kq^2$

```python
import numpy as np
from matplotlib import pyplot as plt

m = 1.0
k = 1.0
eps = 1e-7

fig, ax = plt.subplots()
ax.set_ylabel("p")
ax.set_xlabel("q")
ax.set_aspect(1)
ax.set_title("Harmonic Oscillator Contour Plot")

energies = np.arange(2,20,2)/10

for E in energies:
    q = np.linspace(-np.sqrt(2*E/k) + eps, np.sqrt(2*E/k) - eps, 500)
    
    p_pos = np.sqrt(2*m*(E - 0.5 * k * q**2))
    p_neg = -p_pos

    ax.plot(q,p_pos, color="green")
    ax.plot(q,p_neg, color="green")

plt.show()
```

### Double Well Potential
$U(q) = \frac{D_0}{a^4}(q^2-a^2)^2$

```python
D0 = 1.0
a = 1.0

fig, ax = plt.subplots()
ax.set_ylabel("p")
ax.set_xlabel("q")
ax.set_title("Double Well Contour Plot")

for E in energies:
    if E < D0:
        border_point1 = np.sqrt(a**2 + np.sqrt(a**4 * E/D0))
        border_point2 = np.sqrt(a**2 - np.sqrt(a**4 * E/D0))
        q1 = np.linspace(-border_point1 + eps, -border_point2 - eps, 250)
        q2 = np.linspace(border_point2 + eps, border_point1 - eps, 250)

        p1_pos = np.sqrt(2*m*(E - D0/a**4 * (q1**2 - a**2)**2))
        p1_neg = -p1_pos
        p2_pos = np.sqrt(2*m*(E - D0/a**4 * (q2**2 - a**2)**2))
        p2_neg = -p2_pos

        ax.plot(q1,p1_pos, color="green")
        ax.plot(q1,p1_neg, color="green")
        ax.plot(q2,p2_pos, color="green")
        ax.plot(q2,p2_neg, color="green")

    else:
        border_point = np.sqrt(a**2 + np.sqrt(a**4 * E/D0))
        q = np.linspace(-border_point + eps, border_point - eps, 500)

        p_pos = np.sqrt(2*m*(E - D0/a**4 * (q**2 - a**2)**2))
        p_neg = -p_pos

        
        if E == D0:
            ax.plot(q,p_pos, color="red")
            ax.plot(q,p_neg, color="red")

        else:
            ax.plot(q,p_pos, color="blue")
            ax.plot(q,p_neg, color="blue")

plt.show()
```
